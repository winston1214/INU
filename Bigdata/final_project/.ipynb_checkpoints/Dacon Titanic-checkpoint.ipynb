```python
def show_height(ax,fontsize=10):
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2., height + 0.1,height ,ha="center",fontsize=fontsize)

import os
import matplotlib.font_manager as fm
from matplotlib import rc

def set_font(font_file='malgun.ttf'): # 한글 출력
    font_dir = 'c:/Windows/Fonts/'
    font_name = fm.FontProperties(fname=os.path.join(font_dir, font_file)).get_name()
    rc('font', family=font_name)
set_font()
```

# Modeling

### train test split

```python
from sklearn.model_selection import train_test_split
train,val = train_test_split(train,test_size = .2,random_state = 42)
```

## Automl

```python
from pycaret.classification import *
```

```python
clf = setup(data = train, target = 'survived')
```

```python
best_5 = compare_models(sort = 'AUC', n_select = 5)
```

```python
blended = blend_models(estimator_list = best_5, fold = 5, method = 'soft') # soft voting
```

```python
pred_holdout = predict_model(blended)
```

```python
final_model = finalize_model(blended)
```

```python
val = val.reset_index(drop=True)
val_x = val.drop('survived',axis=1)
val_y = val.survived
```

```python
predictions = predict_model(final_model, data = val_x)
```

```python
predictions
```

```python
from sklearn.metrics import roc_auc_score
validation_auc = roc_auc_score(val_y,predictions['Score'])*100
print('Validation AUC score: %.2f'%validation_auc)
```

## Test set AutoML

```python
predictions_test = predict_model(final_model, data = test)
```

```python
predictions_test
```

```python
submission['Survived'] = predictions_test['Score']
submission.to_csv('automl_test.csv',index=False) # test set auc score : 82.63
```

## Tunning Modeling

```python
cat = create_model('catboost', cross_validation = False)
```

```python
lgbm = create_model('lightgbm', cross_validation = False)
```

```python
cat_tuning = tune_model(cat, optimize = 'AUC', n_iter = 100) ## 평균 88로 성능향상을 보임
```

```python
lgbm_tuning = tune_model(lgbm, optimize = 'AUC', n_iter = 100) # lgbm의 튜닝은 기존의 auc score보다 줄어듦
```

```python
## 따라서 기존의 best_5모델에 voting 값을 추가하고, cat_tunning 모델을 추가한다.
```

```python
best_5.append(cat_tuning)
best_5.append(blended)
```

```python
final_tunning_model = blend_models(estimator_list = best_5, fold = 5, method = 'soft')
```

```python
tunning_model_predict = predict_model(final_tunning_model, data = val_x)
```

```python
print(roc_auc_score(val_y,tunning_model_predict['Score'])*100)
```
