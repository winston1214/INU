```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data,columns = iris.feature_names)
```

```python
df['Target'] = iris.target
```

```python
df.columns
```

# 1ë²ˆ

```python
sns.scatterplot(df['sepal length (cm)'],df['sepal width (cm)'],hue=df['Target'])
```

```python
sns.scatterplot(df['sepal length (cm)'],df['petal length (cm)'],hue=df['Target'])
```

# 2ë²ˆ

```python
mean_df = df.groupby('Target').mean()
mean_df.T
```

# 3ë²ˆ

```python
from sklearn.model_selection import train_test_split
train,test = train_test_split(df,test_size=0.25)
```

```python
train.to_csv('train_dataset.csv',index=False)
test.to_csv('test_dataset.csv',index=False)
```
