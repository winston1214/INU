# 2번

```python
import numpy as np
onehots={}
onehots['cat'] = [1,0,0,0]
onehots['the']=[0,1,0,0]
onehots['dog'] = [0,0,1,0]
onehots['sat'] = [0,0,0,1]
sentence=['the','cat','sat']
x=np.array([0,0,0,0])
for i in onehots.keys():
    if i in sentence:
        x+=np.array(onehots[i])
print("Sentence Encoding: ",str(x))
```

# 3번

```python
# 파란줄
x=3
y=-4
z=2
w=1
q=x*y
p=w+z
f = q*p
print('q={},p={},f={}'.format(q,p,f))


dqdx = y # q x로 미분
dqdy= x # q y로 미분
dpdz = 1 # p를 z로 미분
dpdw = 1 # p를 w로 미분

dfdq = p # f q로 미분
dfdp = q # f p로 미분

dfdx = dfdq*dqdx # 연쇄법칙
dfdy = dfdq * dqdy# 연쇄법칙
dfdz = dfdp*dpdz # 연쇄법칙
dfdw = dfdp*dpdw # 연쇄법칙

print("dfdx:{} , dfdy: {}, dfdz: {}, dfdw: {}".format(dfdx,dfdy,dfdz,dfdw))
print("dfdq:{}, dfdp:{}".format(dfdq,dfdp))
```

# 4-1

```python
def activation(x):
    return 2*x
```

```python
input_x = np.array([1,0.5]) # 입력값
input_w=np.array([[0.8,0.4],[0.9,0.6]]) # 입력값에서 아웃풋으로 가는 가중치
output = np.dot(input_w,input_x) # 가중치와 입력값 행렬곱
output = activation(output) # 활성화 함수 적용
output
```

# 4-2

```python
target=np.array([1.2,1.9]) # 목표값
error = abs(target-output) # error의 절대값
w=input_w
normalize = np.array([[w[0,0]/(w[0,0]+w[1,0]),w[0,1]/(w[0,1]+w[1,1])],
                 [w[1,0]/(w[1,0]+w[0,0]),w[1,1]/(w[0,1]+w[1,1])]]) # 정규화
```

```python
hidden_error = np.dot(normalize,error)
hidden_error
```
