## Bert 돌린 이후

```python
samsung_news1 = pd.read_excel('naver_news/naver_삼성전자_result_0101_1031.xlsx',parse_dates = ['date'])
label = pd.read_csv('bert_data/samsung_news1.csv')['pred']
```

```python
samsung_news1 = samsung_news1[['title','date']]
samsung_news1['label'] = label
```

```python
samsung_news2 =  pd.read_excel('naver_news/naver_삼성전자_result_0101_1031_2.xlsx',parse_dates = ['date'])
label = pd.read_csv('bert_data/samsung_news2.csv')['pred']
```

```python
samsung_news2 = samsung_news2[['title','date']]
samsung_news2['label'] = label
```

```python
up_idx = []
down_idx = []
for i in samsung_news1['label']:
    if i == 1:
        up_idx.append(1)
        down_idx.append(0)
    else:
        up_idx.append(0)
        down_idx.append(1)
samsung_news1['up_count'] = up_idx
samsung_news1['down_count'] = down_idx
```

```python
samsung_news1_pivot = pd.pivot_table(samsung_news1,index = 'date',values = ['up_count','down_count'],aggfunc = 'sum').reset_index()
```

```python
def make_pivot(data_name,label_name):
    data = pd.read_excel('naver_news/'+data_name,parse_dates = ['date'])
    label = pd.read_csv('bert_data/'+label_name)['pred']
    data = data[['title','date']]
    data['label'] = label
    up_idx = []
    down_idx = []
    for i in data['label']:
        if i == 1:
            up_idx.append(1)
            down_idx.append(0)
        else:
            up_idx.append(0)
            down_idx.append(1)
    data['up_count'] = up_idx
    data['down_count'] = down_idx
    pivot = pd.pivot_table(data,index = 'date',values = ['up_count','down_count'],aggfunc = 'sum').reset_index()
    return pivot
```

```python
samsung_news2_pivot = make_pivot('naver_삼성전자_result_0101_1031_2.xlsx','samsung_news2.csv')
```

```python
samsung_news_bert = pd.merge(samsung_news1_pivot,samsung_news2_pivot,on='date',how='outer')
samsung_news_bert.fillna(0,inplace=True)
```

```python
samsung_news_bert['up_count'] = samsung_news_bert['up_count_x'] + samsung_news_bert['up_count_y']
samsung_news_bert['down_count'] = samsung_news_bert['down_count_x'] + samsung_news_bert['down_count_y']
```

```python
samsung_news_bert.drop(['down_count_x','up_count_x','up_count_y','down_count_y'],axis=1,inplace = True)
```

```python
samsung_news_bert.to_csv('bert_data/final_bert_samsung.csv',index=False)
```

```python
def make_pivot_test(data_name,label_name):
    data = pd.read_excel('test_data/'+data_name,parse_dates = ['date'])
    label = pd.read_csv('bert_data/test/'+label_name)['pred']
    data = data[['title','date']]
    data['label'] = label
    up_idx = []
    down_idx = []
    for i in data['label']:
        if i == 1:
            up_idx.append(1)
            down_idx.append(0)
        else:
            up_idx.append(0)
            down_idx.append(1)
    data['up_count'] = up_idx
    data['down_count'] = down_idx
    pivot = pd.pivot_table(data,index = 'date',values = ['up_count','down_count'],aggfunc = 'sum').reset_index()
    return pivot
```

```python
samsung_test_pivot = make_pivot_test('naver_삼성전자_result_1101_1115.xlsx','samsung_news_test_bert1.csv')
samsung_test_pivot2 = make_pivot_test('naver_삼성전자_result_1101_1115_2.xlsx','samsung_news_test_bert2.csv')
```

```python
samsung_test_bert = pd.merge(samsung_test_pivot,samsung_test_pivot2,on='date',how='outer')
samsung_test_bert.fillna(0,inplace=True)
samsung_test_bert
```

```python
samsung_test_bert['up_count'] = samsung_test_bert.up_count_x + samsung_test_bert.up_count_y
samsung_test_bert['down_count'] = samsung_test_bert.down_count_x + samsung_test_bert.down_count_y
samsung_test_bert.drop(['down_count_x','up_count_x','up_count_y','down_count_y'],axis=1,inplace=True)
```
