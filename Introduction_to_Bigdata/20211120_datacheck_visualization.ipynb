```python
def stock_plot(data,date_col,price_col,save_name,title_name):
    plt.figure(figsize = (20,12))
    # plt.plot(samsung_merge1['date'],samsung_merge1['samsung_price'])
    sns.lineplot(data[date_col],data[price_col])
    plt.title(f'{title_name}',fontsize=15)
    plt.xticks(rotation = 45)
    plt.xlabel('date',fontsize=15)
    plt.ylabel('stock',fontsize = 15)
    plt.savefig(f'{save_name}.png')
    plt.show()
```

```python
def error_date_find(data,col):
    korean = re.compile('[\u3131-\u3163\uac00-\ud7a3]+')
    error_date = []
    for idx, v in enumerate(data[col]):
        kor_find = re.findall(korean, v)
        if kor_find : 
            error_date.append(idx)
    return error_date
```

# Modeling

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state = 42)
```

```python
train = samsung_merge1[['samsung_price','samsung_change']]
```

```python
test = pd.read_csv('test_data/samsung_price.csv')
```

```python
test = test[['종가','변동 %']]
```

```python
test.columns = ['samsung_price','samsung_change']
```

```python
train['samsung_change'] = train['samsung_change'].apply(lambda x: x[:-1])
```

```python
remove_idx = []
for i,v in enumerate(train['samsung_change']):
    if v == '0.00':
        remove_idx.append(i)
```

```python
train.drop(remove_idx,axis=0,inplace=True)
```

```python
train['samsung_change'] = train['samsung_change'].astype(float)
train['samsung_change'] = train['samsung_change'].apply(lambda x: 1 if x>0 else 0)
```

```python
test['samsung_change'] = test['samsung_change'].apply(lambda x: x[:-1])
remove_idx = []
for i,v in enumerate(test['samsung_change']):
    if v == '0.00':
        remove_idx.append(i)
test.drop(remove_idx,axis=0,inplace=True)
test['samsung_change'] = test['samsung_change'].astype(float)
test['samsung_change'] = test['samsung_change'].apply(lambda x: 1 if x>0 else 0)
```

```python
test['samsung_price'] = test['samsung_price'].apply(lambda x: int(x.replace(',','')))
```

```python
train_x = train['samsung_price'].values.reshape(-1,1)
train_y = train['samsung_change'].values.reshape(-1,1)
test_x = test['samsung_price'].values.reshape(-1,1)
test_y = test['samsung_change'].values.reshape(-1,1)
```

```python
from sklearn.metrics import accuracy_score
rf.fit(train_x,train_y)
accuracy_score(test_y,rf.predict(test_x))
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(random_state = 42)
xgb.fit(train_x,train_y)
accuracy_score(test_y,xgb.predict(test_x))
```
